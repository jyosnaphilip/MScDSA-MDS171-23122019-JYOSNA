develop  a contact management program where users cAN ADD, edit delete and search contact details usig a dictionary
menu options would facilitate these actions.


```python

Contact={}
def addContact():
    n=input("enter contact name:")
    p=input("enter phone number:")
    n=n.strip().title()
    p=p.strip()
    if n in Contact.keys():
        print("contact already exists")
    
    else:
        Contact.update({n:p})
        print("Contact added successfully!")

def editContact():
    n=input("enter contact name to be edited:")
    if n in Contact.keys():
        print("a. edit name")
        print("b. edit number")
        o=input("enter your choice:")
        if o=="a":
            new_N=input("enter the updated name:")
            Contact.update({new_N:Contact[n]})
            Contact.pop(n)
            print("Contact edited successfully!")
        
        elif o=="b":
            new_P=input("enter updated number:")
            Contact[n]=new_P
            print("phone number edited successfully")
        else:
            print("invalid entry")
    else:
        print("name doesn't exist!")

def delContact():
    c=input("enter contact name to be deleted:")
    if c in Contact.keys():
        Contact.pop(c)
        print("contact deleted successfully!")
    else:
        print("name doesnt exist!")

def searchContact():
    print("a.search by name")
    print("b.search by number")
    o=input("enter your choice:")
    if o=="a":
        name=input("enter name:")
        if name in Contact.keys():
            print(name,":",Contact[name])
        else:
            print("name doesnt exist")
    elif o=="b":
        p=input("enter phone number:")
        if p in Contact.values():
            print(Contact.get(p))
        else:
            print("number not found!")
    else:
        print("invalid option")
while True:
    print("CONTACTS")
    print("*"*30)
    print("1. add a contact")
    print("2. edit a contact")
    print("3. delete a contact")
    print("4. search a contact")
    print("5.exit")
    option=int(input("enter your option:"))
    if option==1:
        addContact()
    elif option==2:
        editContact()
    elif option==3:
        delContact()
    elif option ==4:
        searchContact()
    elif option==5:
        exit()
    else:
        print("not a valid option !")
```
