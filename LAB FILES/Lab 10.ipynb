Create a Python Class for managing student details and marks.
Define the class and implement the methods of the student class in a menu-driven program for different types of users


```python
class studentDetails:
    def __init__(self):
        self.students_dict={}
    def addStudent(self,s_roll,s_name,emailid):
        self.students_dict[s_roll]={"name":s_name,"email_id":emailid}
    def updateDetails(self,s_roll,detail,new_detail):
        for student_roll in self.students_dict:
            if student_roll == s_roll:
                self.students_dict[student_roll][detail]=new_detail
        
    def addMarks(self,regno,subject,marks):
        self.students_dict[regno][subject]=marks



    def displayStudent(self):
        
        for student in self.students_dict:
            print("reg_no:",student)
            for key in self.students_dict[student]:
                print(key,":",self.students_dict[student][key])
            print("*"*10)



        
    def searchStudent(self,regno):
        if regno in self.students_dict:
            print("reg_no:",regno)
            for key in self.students_dict[regno]:
                print(key,":",self.students_dict[regno][key])
            print("*"*10)
        else:
            print("regno not found")
        

            
```

```python
msds=studentDetails()
```

```python
while True:
    print("1.student")
    print("2.teacher")
    print("3.exit")
    option=input("enter option:")
    if option=="1":
        print("a. View details")
        print("b.exit")
        student_choice=input("enter option:")
        if student_choice=="a":
            student_regno=input("enter your regno:")
            msds.searchStudent(student_regno)
        
        elif student_choice=="b":
            break

        else:
            print("invalid entry")

    elif option=="2":
        print("a.add student")
        print("b.update student details")
        print("c.add marks")
        print("d. search a student")
        print("e.display students")
        print("f.exit")
        teacher_choice=input("enter option:")

        if teacher_choice == "a":
            name=input("enter name:")
            regno=input("enter regno")
            emailid=input("enter emailid")
            msds.addStudent(regno,name,emailid)
        
        elif teacher_choice == "b":
            regno=input("enter regno:")
            msds.searchStudent(regno)
            detail=input("What detail do you want to change")
            new_detail=input("enter new detail:")
            msds.updateDetails(regno,detail.lower().strip(),new_detail.lower().strip())
       
        elif teacher_choice == "c":
            regno=input("enter reg no:")
            subject=input("enter subject")
            mark=input("enter marks:")
            msds.addMarks(regno,subject.lower().strip(),mark)
        elif teacher_choice == "d":
            student_regno=input("enter student regno:")
            msds.searchStudent(student_regno)

        elif teacher_choice == "e":
            msds.displayStudent()
        elif teacher_choice == "f":
            break
        else:
            print("invalid entry")
    
    elif option=="3":
        break

    else:
        print("invalid entry")
    
```
